```python
class Node:
  def __init__(self,data):
    self.data = data
    self.pref = None
    self.nref = None
class DoublyLinkedList:
  def __init__(self):
    self.head = None
  def printLL(self):
    if self.head is None:
      print("DLL is empty.")
      print("\n")
    else:
      n = self.head
      while n is not None:
        print(n.data,"-->",end=" ")
        n = n.nref
      print("\n")
  def insertAtBeginning(self,data):
    if self.head is None:
      new_node = Node(data)
      new_node.nref = self.head
      self.head = new_node
    else:
      new_node = Node(data)
      new_node.nref = self.head
      self.head.pref = new_node
      self.head = new_node
  def insertAtEnd(self,data):
    if self.head is None:
      print("Linked List is empty.")
    else:
      new_node = Node(data)
      n = self.head
      while n.nref is not None:
        n = n.nref
      n.nref = new_node
      new_node.pref = n
  def insertAtBetween(self,data,x):
    if self.head is None:
      print("DLL is empty.")
    else:
      n = self.head
      while n.data is not x:
        n = n.nref
      new_node = Node(data)
      temp = n.nref
      n.nref = new_node
      new_node.pref = n
      new_node.nref = temp
      temp.pref = new_node
  def deleteFromBeginning(self):
    if self.head is None:
      print("DLL can't be deleted")
    else:
      temp = self.head.nref
      self.head.nref = None
      self.head = temp
  def deleteFromEnd(self):
    if self.head is None:
      print("DLL can't be deleted")
    else:
      n = self.head
      while n.nref.nref is not None:
        n =n.nref
      n.nref = None
  def deleteFromBetween(self,x):
    if self.head is None:
      print("DLL is empty.")
    else:
      n = self.head
      while n.data is not x:
        n = n.nref
      temp = n.nref.nref
      n.nref = n.nref.nref
      temp.pref = n
DLL1 = DoublyLinkedList()
DLL1.insertAtBeginning(10)
DLL1.insertAtBeginning(20)
DLL1.insertAtBeginning(30)
DLL1.insertAtBeginning(40)
DLL1.printLL()
DLL1.insertAtEnd(50)
DLL1.insertAtEnd(60)
DLL1.printLL()
DLL1.insertAtBetween(70,30)
DLL1.insertAtBetween(80,50)
DLL1.printLL()
DLL1.deleteFromBeginning()
DLL1.deleteFromBeginning()
DLL1.printLL()
DLL1.deleteFromEnd()
DLL1.deleteFromEnd()
DLL1.printLL()
DLL1.deleteFromBetween(20)
DLL1.deleteFromBetween(70)
DLL1.printLL()
```
